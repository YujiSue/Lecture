2. 「健康診断データ」「新型コロナ検査」以外の診療データをもう１セット作成してカルテに追加したのち、Pythonを実行して結果ファイルを提出すること。(チャレンジャー向け)  
3. 下のコード中の「カルテ」オブジェクトは、講義中に説明した「カルテ」オブジェクトから変更点が２点ある。  
「変更点」とコメントがついた２か所のコードの何がどう変わったのか、コメントに説明を付け足して提出すること。(経験者向け) 

<font color="red">※ファイルの保存方法：左上メニューの「ファイル」から「.ipynbをダウンロード」を選んで、ダウンロード</font>   



```
# 日付を扱う別プログラムを読み込む
import datetime

###############　カルテ【オブジェクト】の作成　##################
class 簡易カルテ:
  # 初期状態の設定
  def __init__(self):
    # 変数の設定
    self.患者ID = None
    self.患者氏名 = None
    self.患者生年月日 = None
    self.診療データリスト = []

  # 患者の情報を記入する関数
  def 記入(self, ID, 名前, 生年月日):
    self.患者ID = ID
    self.患者氏名 = 名前
    self.患者生年月日 = 生年月日
  
  # 変更点１
  # 患者の年齢を計算
  def 年齢(self):
    生年 = int(生年月日.split('-')[0])
    return 2020 - 生年

  # 診療データを追加する関数
  def データ追加(self, データ):
    self.診療データリスト.append(データ)

  # 指定のページを表示する関数
  def 表示(self, ページ):
    if (ページ == 1):
      print('ID:　',self.患者ID)
      print('氏名:　',self.患者氏名)
      print('生年月日:　',self.患者生年月日, '(', self.年齢(), '歳)')
    # 変更点２
    elif (ページ-2 < len(self.診療データリスト)):
      print(self.診療データリスト[ページ-2].データのまとめ())
    else:
      print('まだ', ページ, '頁目はありません')

###############################################################

###############　診療データ【オブジェクト】の作成　##################
class 診療データ:
  # 初期状態の設定
  def __init__(self):
    # 変数の設定
    self.診療名 = ''
    self.診療日 = datetime.date.today()
    self.診療項目 = []
    self.診療結果 = {}
    self.医師の所見 = ''
  
  # 診療情報を記入する関数
  def 診療情報の記入(self, 内容):
    self.診療名 = 内容['名称']
    self.診療項目 = 内容['項目']

  # 診療結果を記入する関数
  def 診療結果の記入(self, 項目, 結果):
    self.診療結果[項目] = 結果

  # 所見を記入する関数
  def 所見の記入(self, 所見):
    self.医師の所見 = 所見

  # 要約する関数
  def データのまとめ(self):
    データ = '－－－－－－－－－－－－－－－－－－－－－\n'
    データ += '診療：'+self.診療名+'　　診療日：'+self.診療日.strftime('%Y年%m月%d日')+'\n'
    データ += '－－－－－－－－－－－－－－－－－－－－－\n'
    データ += '　診療項目　 結果\n'
    for 項目名 in self.診療項目:
      データ += '　　'+項目名+'　　'+self.診療結果[項目名]+'\n'
    データ += '所見：'+self.医師の所見+'\n'
    return データ

###################################################################

# カルテを運用する

# カルテの作成
私のカルテ = 簡易カルテ() 

#@markdown ###私のカルテ
学籍番号 =  11200062#@param
名前 = '\u7389\u91CE\u6075\u90C1' #@param {type:"string"}
生年月日 = '2000-12-12' #@param {type:"date"}
私のカルテ.記入(学籍番号, 名前, 生年月日)

# データの追加１
健康診断データ = 診療データ()
健康診断データ.診療情報の記入({ '名称': '健康診断', '項目': [ '身長', '体重', '視力', '血圧' ] })

#@markdown ###健康診断データ
身長 = '180 cm' #@param
体重 = '70.0 kg' #@param
視力 = '2.0/2.0' #@param
血圧 = '150/80' #@param

健康診断データ.診療結果の記入('身長', 身長)
健康診断データ.診療結果の記入('体重', 体重)
健康診断データ.診療結果の記入('視力', 視力)
健康診断データ.診療結果の記入('血圧', 血圧)
健康診断データ.所見の記入('異常なし')

私のカルテ.データ追加(健康診断データ)


# データの追加２
新型コロナ検査 = 診療データ()
新型コロナ検査.診療情報の記入({ '名称': '新型コロナ検査', '項目': [ '体温', 'せき', '味覚', 'PCR ' ] })

#@markdown ###新型コロナ検査
体温 = '37.5 ℃' #@param
せき = 'なし' #@param
味覚 = '正常' #@param
PCR = '陰性' #@param

新型コロナ検査.診療結果の記入('体温', 体温)
新型コロナ検査.診療結果の記入('せき', せき)
新型コロナ検査.診療結果の記入('味覚', 味覚)
新型コロナ検査.診療結果の記入('PCR ', PCR)
新型コロナ検査.所見の記入('非感染者')

私のカルテ.データ追加(新型コロナ検査)

私のカルテ.表示(1)
私のカルテ.表示(2)
私のカルテ.表示(3)
私のカルテ.表示(4)
```
